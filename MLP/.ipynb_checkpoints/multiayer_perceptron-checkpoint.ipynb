## Multi-Layer Perceptron:
A Multi-Layer Perceptron (MLP) is a kind of artificial neural network used in deep learning and machine learning. Given that it is a feedforward neural network, information moves without loops or cycles from input to output in a single direction. MLPs are frequently utilized for many different applications, including as classification, regression, and pattern recognition.

### Importing the libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### defining Step Function

```python
def step_function(s):
    return np.array(s > 0).astype(int)
```

```python
x = np.arange(-10, 10, 0.1)
y = step_function(x)
```

### Visualizing Step Function

```python
plt.plot(x, y)
plt.title('this is a proud title')
plt.show()
```

### Defining Sigmoid Function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
sigmoid(np.array([-3,-2,-1, 0,1,2,3]))
```

```python
x = np.arange(-9, 9, 0.2)
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.title('sigmoid function')
plt.show()
```

### MLP Implementation

```python
import numpy as np
```

#### Z1

```python
x = np.array([4, 6])
w1 = np.array([[0.2,0.4,0.6],[0.1,0.3,0.5]])
b1 = np.array([0.1,0.2,0.3])
a1 = np.dot(x, w1) + b1
```

```python
z1 = sigmoid(a1)
z1
```

#### Z2

```python
w2 = np.array([[0.2,0.4],[0.3,0.4],[0.3,0.5]])
b2 = np.array([0.1,0.3])
a2 = np.dot(z1, w2) + b2
```

```python
z2 = sigmoid(a2)
z2
```

#### Z3

```python
w3 = np.array([[0.4,0.6],[0.1,0.5]])
b3 = np.array([0.2,0.2])
a3 = np.dot(z2,w3) + b3
z3 = sigmoid(a3)
```
